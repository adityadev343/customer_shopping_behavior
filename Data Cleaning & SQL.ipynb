```python
missing_purchase_by_category = df[df['Purchase Amount (USD)'].isna()]['Category'].value_counts()
print("Missing Purchase Amount (USD) by Category:")
print(missing_purchase_by_category)
print(f"Total missing: {missing_purchase_by_category.sum()}\n")

missing_review_by_category = df[df['Review Rating'].isna()]['Category'].value_counts()
print("Missing Review Rating by Category:")
print(missing_review_by_category)
print(f"Total missing: {missing_review_by_category.sum()}")
```

```python
for col in ['Purchase Amount (USD)', 'Review Rating']:
    df[col] = df.groupby('Category')[col].transform(lambda x: x.fillna(x.median()))
print(df[['Purchase Amount (USD)', 'Review Rating']].isnull().sum())
```

```python
def show_categorical_uniques(df, columns=None):
    if columns is None:
        columns = df.select_dtypes(include='object').columns.tolist()
    for col in columns:
        uniq = sorted(df[col].dropna().astype(str).unique())
        print(f"{col} ({len(uniq)} unique): [ {', '.join(repr(v) for v in uniq)} ]")

show_categorical_uniques(df)
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
import numpy as np
bins = [17, 30, 45, 60, np.inf]
labels = ['18-30', '31-45', '46-60', '61+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=True, include_lowest=True)
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days','frequency_of_purchases']].head()
```

```python
df.to_csv('customer_shopping_behavior_cleaned.csv', index=False)
```

# **Revenue & Sales Performance**

## What is the total revenue generated by male vs female customers?

```python
q1 = conn.execute("""
    SELECT
        gender,
        SUM(purchase_amount) AS total_revenue,
        ROUND(100.0 * SUM(purchase_amount) / SUM(SUM(purchase_amount)) OVER(), 2) AS revenue_pct
    FROM shopping
    GROUP BY gender
    ORDER BY total_revenue DESC;
""").fetchdf()
print(q1)
```

## What is the revenue contribution of each age group?

```python
q2 = conn.execute("""
    SELECT
        age_group,
        SUM(purchase_amount) AS revenue,
        ROUND(100.0 * SUM(purchase_amount) / SUM(SUM(purchase_amount)) OVER(), 2) AS revenue_pct
    FROM shopping
    GROUP BY age_group
    ORDER BY revenue DESC;
""").fetchdf()
print(q2)
```

## Which product categories generate the highest total revenue, and what percentage of total revenue do they contribute?

```python
q3 = conn.execute("""
    WITH total AS (SELECT SUM(purchase_amount) AS total_rev FROM shopping)
    SELECT
        category,
        SUM(purchase_amount) AS revenue,
        ROUND(100.0 * SUM(purchase_amount) / (SELECT total_rev FROM total), 2) AS revenue_pct
    FROM shopping
    GROUP BY category
    ORDER BY revenue DESC;
""").fetchdf()
print(q3)
```

## Which shipping methods generate the highest revenue, order volume, and average order value?

```python
q4 = conn.execute("""
    SELECT
        shipping_type,
        COUNT(*) AS order_volume,
        SUM(purchase_amount) AS total_revenue,
        ROUND(AVG(purchase_amount), 2) AS avg_order_value
    FROM shopping
    GROUP BY shipping_type
    ORDER BY total_revenue DESC;
""").fetchdf()
print(q4)
```

## Which Category × Shipping Type × Subscription Status combinations generate the highest revenue?

```python
q5 = conn.execute("""
    SELECT
        category,
        shipping_type,
        subscription_status,
        SUM(purchase_amount) AS total_revenue,
        COUNT(*) AS order_count
    FROM shopping
    GROUP BY category, shipping_type, subscription_status
    ORDER BY total_revenue DESC
    LIMIT 10;
""").fetchdf()
print(q5)
```

# **Customer Segmentation & Value Analysis**

## Segment customers into New, Returning, and Loyal based on previous purchases and determine the size of each segment.

```python
q6 = conn.execute("""
    WITH seg AS (
        SELECT
            customer_id,
            CASE
                WHEN previous_purchases <= 1 THEN 'New'
                WHEN previous_purchases BETWEEN 2 AND 5 THEN 'Returning'
                ELSE 'Loyal'
            END AS segment
        FROM shopping
    )
    SELECT
        segment,
        COUNT(DISTINCT customer_id) AS customer_count
    FROM seg
    GROUP BY segment;
""").fetchdf()
print(q6)
```

## Do subscribed customers spend more?

```python
q7 = conn.execute("""
    SELECT
        subscription_status,
        COUNT(DISTINCT customer_id) AS customers,
        ROUND(AVG(purchase_amount), 2) AS avg_spend,
        SUM(purchase_amount) AS total_revenue
    FROM shopping
    GROUP BY subscription_status;
""").fetchdf()
print(q7)
```

## Are repeat buyers more likely to subscribe?

```python
q8 = conn.execute("""
    SELECT
        SUM(CASE WHEN previous_purchases > 5 AND subscription_status = 'Yes' THEN 1 ELSE 0 END) AS loyal_subscribers,
        SUM(CASE WHEN previous_purchases > 5 THEN 1 ELSE 0 END) AS loyal_total,
        ROUND(100.0 * SUM(CASE WHEN previous_purchases > 5 AND subscription_status = 'Yes' THEN 1 ELSE 0 END) /
              NULLIF(SUM(CASE WHEN previous_purchases > 5 THEN 1 ELSE 0 END), 0), 2) AS pct_subscribed
    FROM shopping;
""").fetchdf()
print(q8)
```

## Which customer segments contribute the most revenue?

```python
q9 = conn.execute("""
    SELECT
        age_group,
        subscription_status,
        CASE
            WHEN previous_purchases <= 5 THEN 'Low frequency'
            WHEN previous_purchases <= 15 THEN 'Mid frequency'
            ELSE 'High frequency'
        END AS freq_bin,
        SUM(purchase_amount) AS total_revenue
    FROM shopping
    GROUP BY age_group, subscription_status, freq_bin
    ORDER BY total_revenue DESC
    LIMIT 10;
""").fetchdf()
print(q9)
```

##  Which customer groups generate the highest estimated Customer Lifetime Value (CLV)?

```python
q10 = conn.execute("""
    SELECT
        age_group,
        subscription_status,
        ROUND(AVG(purchase_amount * previous_purchases), 2) AS avg_clv
    FROM shopping
    GROUP BY age_group, subscription_status
    ORDER BY avg_clv DESC;
""").fetchdf()
print(q10)
```

## Who are the top 10% highest-spending customers, and what characteristics do they share?

```python
q11 = conn.execute("""
    WITH customer_spend AS (
        SELECT
            customer_id,
            SUM(purchase_amount) AS total_spend,
            PERCENT_RANK() OVER (ORDER BY SUM(purchase_amount)) AS spend_percentile
        FROM shopping
        GROUP BY customer_id
    )
    SELECT
        cs.customer_id,
        cs.total_spend,
        s.gender,
        s.age_group,
        s.subscription_status
    FROM customer_spend cs
    JOIN shopping s ON cs.customer_id = s.customer_id
    WHERE cs.spend_percentile >= 0.9
    GROUP BY cs.customer_id, cs.total_spend, s.gender, s.age_group, s.subscription_status
    ORDER BY cs.total_spend DESC
    LIMIT 20;
""").fetchdf()
print(q11)
```

# **Product Performance Analysis**

## Which are the top 5 products with the highest average review rating (minimum 30 purchases)

```python
q12 = conn.execute("""
    SELECT
        item_purchased,
        COUNT(*) AS num_purchases,
        ROUND(AVG(review_rating), 2) AS avg_rating
    FROM shopping
    GROUP BY item_purchased
    HAVING COUNT(*) >= 30
    ORDER BY avg_rating DESC
    LIMIT 5;
""").fetchdf()
print(q12)
```

## What are the top 3 most purchased products within each category?

```python
q13 = conn.execute("""
    WITH ranked AS (
        SELECT
            category,
            item_purchased,
            COUNT(*) AS purchase_count,
            ROW_NUMBER() OVER (PARTITION BY category ORDER BY COUNT(*) DESC) AS rn
        FROM shopping
        GROUP BY category, item_purchased
    )
    SELECT
        category,
        item_purchased,
        purchase_count
    FROM ranked
    WHERE rn <= 3
    ORDER BY category, purchase_count DESC;
""").fetchdf()
print(q13)
```

## Which products generate the highest total revenue?

```python
q14 = conn.execute("""
    SELECT
        item_purchased,
        SUM(purchase_amount) AS total_revenue
    FROM shopping
    GROUP BY item_purchased
    ORDER BY total_revenue DESC
    LIMIT 10;
""").fetchdf()
print(q14)
```

## Which products are high-revenue but low-rated, indicating possible customer satisfaction issues?

```python
q15 = conn.execute("""
    SELECT
        item_purchased,
        SUM(purchase_amount) AS total_revenue,
        ROUND(AVG(review_rating), 2) AS avg_rating
    FROM shopping
    GROUP BY item_purchased
    HAVING SUM(purchase_amount) > 5000 AND AVG(review_rating) < 3.5
    ORDER BY total_revenue DESC;
""").fetchdf()
print(q15)
```

# **Discount & Promotion Effectiveness**

## Which customers used a discount but still spent more than the overall average purchase amount?

```python
q16 = conn.execute("""
    WITH overall_avg AS (SELECT AVG(purchase_amount) AS avg_all FROM shopping)
    SELECT
        customer_id,
        SUM(purchase_amount) AS total_spent,
        COUNT(*) AS transactions
    FROM shopping
    WHERE discount_applied = 'Yes'
    GROUP BY customer_id
    HAVING SUM(purchase_amount) > (SELECT avg_all FROM overall_avg)
    ORDER BY total_spent DESC
    LIMIT 10;
""").fetchdf()
print(q16)
```

## Which 5 products have the highest percentage of purchases with discounts applied?

```python
q17 = conn.execute("""
    SELECT
        item_purchased,
        COUNT(*) AS total_purchases,
        SUM(CASE WHEN discount_applied = 'Yes' THEN 1 ELSE 0 END) AS discount_purchases,
        ROUND(100.0 * SUM(CASE WHEN discount_applied = 'Yes' THEN 1 ELSE 0 END) / COUNT(*), 2) AS discount_pct
    FROM shopping
    GROUP BY item_purchased
    HAVING COUNT(*) >= 10
    ORDER BY discount_pct DESC
    LIMIT 5;
""").fetchdf()
print(q17)
```

## How does discount usage affect average purchase amount across categories and shipping methods ?

```python
q18 = conn.execute("""
    SELECT
        category,
        shipping_type,
        ROUND(AVG(CASE WHEN discount_applied = 'Yes' THEN purchase_amount ELSE NULL END), 2) AS avg_with_discount,
        ROUND(AVG(CASE WHEN discount_applied = 'No' THEN purchase_amount ELSE NULL END), 2) AS avg_without_discount,
        COUNT(*) AS total_orders
    FROM shopping
    GROUP BY category, shipping_type
    HAVING COUNT(*) >= 10
    ORDER BY (avg_with_discount - avg_without_discount) DESC;
""").fetchdf()
print(q18.head(10).to_string(index=False))
```

## Which product categories are most dependent on discounts for revenue generation?

```python
q19 = conn.execute("""
    SELECT
        category,
        SUM(CASE WHEN discount_applied = 'Yes' THEN purchase_amount ELSE 0 END) AS discount_revenue,
        SUM(purchase_amount) AS total_revenue,
        ROUND(100.0 * SUM(CASE WHEN discount_applied = 'Yes' THEN purchase_amount ELSE 0 END) / SUM(purchase_amount), 2) AS revenue_pct_from_discount
    FROM shopping
    GROUP BY category
    ORDER BY revenue_pct_from_discount DESC;
""").fetchdf()
print(q19)
```

# **Logistics & Operational Insights**

## Compare average purchase amounts between Standard and Express Shipping.

```python
q20 = conn.execute("""
    SELECT
        shipping_type,
        ROUND(AVG(purchase_amount), 2) AS avg_purchase,
        COUNT(*) AS orders
    FROM shopping
    WHERE shipping_type IN ('Standard', 'Express')
    GROUP BY shipping_type;
""").fetchdf()
print(q20)
```

## Which shipping methods are associated with higher-value purchases?

```python
p90_threshold = conn.execute("SELECT quantile_cont(purchase_amount, 0.9) FROM shopping").fetchone()[0]

q21 = conn.execute(f"""
    SELECT
        shipping_type,
        COUNT(*) AS high_value_orders,
        ROUND(AVG(purchase_amount), 2) AS avg_high_value
    FROM shopping
    WHERE purchase_amount >= {p90_threshold}
    GROUP BY shipping_type
    ORDER BY high_value_orders DESC;
""").fetchdf()
print(q21)
```
